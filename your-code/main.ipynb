# Challenge 1 - Fork Languages

You will find out how many programming languages are used among all the forks created from the main lab repo of ironhack Madrid. The URL is  `ironhack-datalabs/madrid-oct-2018`, you will:


1. Obtain the full list of forks created from the main lab repo via Github API.

1. Loop the JSON response to find out the `language` attribute of each fork. Use an array to store the `language` attributes of each fork.
    * *Hint: Each language should appear only once in your array.*

1. Print the language array. It should be something like:

	```["Python", "Jupyter Notebook", "HTML"]``` 
			or
	```["Jupyter Notebook", "HTML", None]```

Again, the documentation of Github API is [here](https://developer.github.com/v3/).

````python
import pandas as pd
import requests
````

````python
url = "https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018"
response = requests.get(url=url)
response.status_code
````

````python
iron_madrid = response.json()
iron_madrid
````

````python
for key, value in iron_madrid.items():
    print(key,"-->", value)
````

````python
url = "https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/forks"
response = requests.get(url=url)
response.status_code
iron_madrid = response.json()
iron_madrid
````

````python
list_languages = []

for dict in iron_madrid:
        if dict['language'] not in list_languages:
                list_languages.append(dict['language'])
        else:
                pass

list_languages
````

````python
url = "https://api.github.com/repos/IAndradeDA/lab-list-comprehensions/commits"

response = requests.get(url=url, params={'since':'str(2022-10-09)', 'until':'str(2022-10-23)'})
response.status_code
````

````python
response.json()
````

````python
commi = pd.DataFrame(response.json())
commi
````

````python
print("The number of commits I made in the past 2 weeks with that lab were: ", commi['commit'].value_counts().sum())
````

````python
for dict in commi['commit']:
    print(dict)
````
